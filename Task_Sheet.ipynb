```python
import re
import datetime
class login_system:
  def sign_up():
    All_valid = True
    user_id = input("enter your userid: ")
    pattern = '\d{3}'
    if re.fullmatch(pattern, user_id):
      print("Valid user_id")
    else:
      print("Invalid user_id")
      All_valid = False
    username = input("Enter your username: ")
    pattern = '[a-zA-Z]+'
    if re.fullmatch(pattern, username):
      print("Valid username")
    else:
      print("Invalid username")
      All_valid = False
    dob= input("enter you dob: ")
    pattern = r'^\d{2}[-/\.]\d{2}[-/\.]\d{4}$'
    if re.fullmatch(pattern, dob):
        print("valid format")
    else:
        print ("invalid format")
        All_valid = False
    email = input("Enter your email: ")
    pattern = '[a-zA-Z0-9._]+@gmail\.com'
    if re.fullmatch(pattern, email):
      print("Valid email")
    else:
      print("Invalid email")
      All_valid = False
    phone = input("Enter your phone number: ")
    pattern = '\d{10}'
    if re.fullmatch(pattern, phone):
      print("Valid phone number")
    else:
      print("Invalid phone number")
      All_valid = False
    password = input("Enter your password: ")
    pattern = r'^(?=.*[A-Z])(?=.*[a-z])(?=.*\d)(?=.*[@$!#%*?&])[A-Za-z\d@$!#%*?&]{8,}$'
    if re.fullmatch(pattern, password):
        print("valid password")
    else:
        print("invalid password")
        All_valid = False
    confirm_password = input("Re-enter your password: ")
    if password == confirm_password:
      print("Password matched")
    else:
      print("Password not matched")
      All_valid = False
    if All_valid:
     print("Registration successfull")
     with open("user.txt", "w") as file:
       file.write(f'user_id|username|dob|email|phone|password\n')
       file.write(f'{user_id}|{username}|{dob}|{email}|{phone}|{password}\n')
    else:
     print("Registration failed")

  def login():
        try:
            # Open the file and read all user details
            with open("user.txt", "r") as file:
                lines = file.readlines()

            # Skip the header and store user credentials in a dictionary
            credentials = {}
            for line in lines[1:]:  # Skip the first line (header)
                user_data = line.strip().split("|")
                if len(user_data) == 6:
                    username = user_data[1]
                    password = user_data[5]
                    credentials[username] = password

            # Ask user to log in
            username = input("Enter your username: ")
            password = input("Enter your password: ")

            # Check credentials
            if username in credentials and credentials[username] == password:
                login_time = datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
                print(f"Login successful! Welcome back, {username}. Last login: {login_time}")

                # Optionally log the login
                with open("user.txt", "a") as log_file:
                    log_file.write(f"{username} logged in at {login_time}\n")
            else:
                print("Invalid username or password. Please try again.")

        except FileNotFoundError:
            print("User data file not found. Please sign up first.")

  def forgot_password():
    username = input("Enter your username: ")

    try:
        with open("user.txt","r") as file:
         for line in file:
          stored_user_is,stored_username,stored_dob,stored_email,stored_phone,stored_password = line.strip().split("|")
          if stored_username == username:
            print(f'Your password is :{stored_password}')
            return
         else:
          print("User not found")


    except FileNotFoundError:
        print("User data file not found. Please ensure the file exists.")

  def update_password():
     username = input("Enter your username: ")

    # Open the file containing user credentials
     try:
        with open("user.txt", "r") as file:
            lines = file.readlines()

        # Search for the user in the file
        for i, line in enumerate(lines):
            stored_user_id,stored_username,stored_dob,stored_phone,stored_email,stored_password = line.strip().split("|")
            if stored_username == username:
                # Prompt for the old password
                old_password = input("Enter your old password: ")

                # Verify the old password
                if stored_password == old_password:
                    # Prompt for the new password
                    new_password = input("Enter your new password: ")

                    # Validate the new password (ensure it's at least 8 characters long)
                    if len(new_password) >= 8:
                        # Update the password in the file
                        lines[i] = f"{stored_user_id}|{stored_username}|{stored_dob}|{stored_phone}|{stored_email}|{new_password}\n"

                        # Write the updated lines back to the file
                        with open("user.txt", "w") as file:
                            file.writelines(lines)

                            print("Password updated successfully!")

                    else:
                        print("New password must be at least 8 characters long.")

                else:
                    print("Old password is incorrect.")


        # If username not found, inform the user
        print("Username not found.")
     except FileNotFoundError:
        print("User data file not found. Please ensure the file exists.")

def main_menu():
    while True:
        print("\n====== Welcome to Login System ======")
        print("1. Sign Up")
        print("2. Login")
        print("3. Forgot Password")
        print("4. Update Password")
        print("5. Exit")

        choice = input("Choose an option (1-5): ")

        if choice == '1':
            login_system.sign_up()
        elif choice == '2':
            login_system.login()
        elif choice == '3':
            login_system.forgot_password()
        elif choice == '4':
            login_system.update_password()
        elif choice == '5':
            print("Thank you! Exiting...")
            break
        else:
            print("Invalid choice. Please select from 1 to 5.")

# Call the menu
main_menu()
```
